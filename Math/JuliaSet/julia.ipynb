## Mandelbrot

```python
mandelbrot = juliaset(0,mandelbrot=True,posy=(-1.5,1.5),posx=(-2,1.2),display=True,precision=3000)
```

## Some examples of Julia Sets

```python
julia1 = juliaset(0.2+0.72j,display=True,precision=1500,colormap='viridis')
```

```python
julia2 = juliaset(0.43 + 0.1j,display=True,precision=2000,colormap='Spectral')
```

```python
julia3 = juliaset(-0.8 + 0.156j,display=True,precision=2000,colormap='viridis',n_iter=80)
```

```python
julia4 = juliaset(0.43 - 0.23j,display=True,precision=2000,colormap='Spectral')
```

```python
julia5 = juliaset(0.43,display=True,precision=2000,colormap='Spectral')
```
