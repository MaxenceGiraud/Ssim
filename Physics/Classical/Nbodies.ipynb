```python
from NBody import Body,Universe

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Defining Bodies and Universe

a = Body('a',1000,[0,0],[2,0])
b = Body('b',1000,[0,100],[2,0])
#c = Body(1000,[0,-1000],[4,0])

bodies = [a,b]

u = Universe(bodies)
```

```python
# Do the simulation
u.update_(1000,0.1)
```

```python
# Plot the trajectories of the bodies

for body in bodies :
    plt.plot(np.array(body.pos_history)[:,0],np.array(body.pos_history)[:,1],label=body.name)

plt.legend()
plt.show()
```
