## Example of Julia Sets with $z_{n+1} = z_n^3 +c$

```python
f3 = lambda z,c : z**3 + c 
```

```python
juliaz3 = juliaset(-0.5+0.5j,precision=1500,f=f3)
```

* ${ z^{2}/2+z^{4}/4+c}$

```python
f2244 = lambda z,c : z**2/2+z**4/4+c
```

```python
j2244 = juliaset(0.7+1.3j,precision=2000,n_iter=150,f=f2244,colormap='Greys')
```

* $1-z^{2}+z^{5}/(2+4z)+c$

```python
fl = lambda z,c : 1-z**2+z**5/(2+4*z)+c
```

```python
jl = juliaset(-0.163+0.085j,precision=2000,n_iter=150,f=fl,colormap='cividis',posx=(-2.5,2.5))
```

## Newton Fractal, i.e. with f of the form $f(z) = z - \frac{g(z)}{g^{'}(z)}$
* $g(z) = z^3 -1$ and so $z_{n+1}=\frac{(2 z_{n}^{3}+1)} {(3 z_{n}^{2}}$

```python
f_n1 = lambda z,c : (2*z**3+1)/(3*z**2)
```

```python
newton = juliaset(1,display=True,precision=2000,n_iter=200,f=f_n1,colormap='viridis')
```
