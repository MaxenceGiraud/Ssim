## Comparaison performances with vectorized code

```python
n_runs = 100
time_nonvec  = time.time()
for _ in range(n_runs):
    k =koch_snowflake(6,vectorize=False,display=False)
time_nonvec  =  (time.time() - time_nonvec) /n_runs

time_vec = time.time()
for _ in range(n_runs):
     k =koch_snowflake(6,display=False)
time_vec  =  (time.time() - time_vec) /n_runs

print("Time non vetorized code :", time_nonvec)
print("Time vetorized code :", time_vec)
print ("Speed up :", time_nonvec/time_vec)
```
